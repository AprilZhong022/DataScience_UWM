```python
import pandas as pd
from pandas import Series, DataFrame
import csv
import math
```

```python
cityData = pd.read_csv('https://raw.githubusercontent.com/tylerharter/s19-city/master/data/spend.csv')
cityData

```

```python
parks = cityData['Agency'] == '51 -  PARKS'
parks = cityData[parks]
parks
```

```python
parksRevenue = parks['Account Type'] == 'Revenue'
parksRevenue = parks[parksRevenue]
parksRevenue
```

```python
parksExpense = parks['Account Type'] == 'Expense'
parksExpense = parks[parksExpense]
parksExpense
```

```python
rev2015 = abs(parksRevenue['2015'].sum())
rev2016 = abs(parksRevenue['2016'].sum())
rev2017 = abs(parksRevenue['2017'].sum())
rev2018 = abs(parksRevenue['2018'].sum())
exp2015 = abs(parksExpense['2015'].sum())
exp2016 = abs(parksExpense['2016'].sum())
exp2017 = abs(parksExpense['2017'].sum())
exp2018 = abs(parksExpense['2018'].sum())

exp2018
```

```python
revlst = [rev2015, rev2016, rev2017, rev2018]
explst = [exp2015, exp2016, exp2017, exp2018]
```

```python
rev_and_exp = {'Revenue': revlst, "Expenses": explst}
rev_and_exp = DataFrame(rev_and_exp)
revEx = rev_and_exp
years = Series([2015,2016,2017,2018])
revEx = revEx.set_index(years)
revEx
```

```python
revExPlot = revEx.plot.bar(stacked=True)
```

```python
revdf = DataFrame({"Revenue": revlst})
revdf = revdf.set_index(years)
expdf = DataFrame({"Expenses": explst})
expdf = expdf.set_index(years)
```

```python
revPlot = revdf.plot.line()
revPlot.set_xticks([2015, 2016, 2017, 2018])
expPlot = expdf.plot.line()
expPlot.set_xticks(years)

None
```

```python
import sqlite3
cityData = cityData.rename(columns={'Account Type': 'Account_Type', 'Character Code': 'Character_Code', '2015': 'year2015',
                                   '2016': 'year2016', '2017': 'year2017', '2018': 'year2018'})
conn = sqlite3.connect('cityData.db')
cityData.to_sql("city_table", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("""select *, SUM(year2016) as sum
from city_table 
where Agency = '51 -  PARKS'
and Account_Type = "Revenue"

""", conn)
```

```python
consistent_parks_expenses = pd.read_sql("""select * from city_table
where Agency = '51 -  PARKS'
and Account_Type = "Expense"
and (Character_Code LIKE '%supplies' 
or Character_Code LIKE '%salaries'
or Character_Code LIKE '%benefits'
or Character_Code LIKE '%inter depart charges'
or Character_Code LIKE '%purchased services')



""", conn)
consistent_parks_expenses
```

```python
cpe_conn = sqlite3.connect('const_exp.db')
consistent_parks_expenses.to_sql("cpe_table", cpe_conn, if_exists="replace", index=False)
```

```python
parks_supplies = pd.read_sql("""select * , sum(year2015) as sum2015, sum(year2016) as sum2016,
sum(year2017) as sum2017, sum(year2018) as sum2018 from cpe_table
where Character_Code LIKE '%supplies'

""", cpe_conn)
```

```python
supplies2015 = parks_supplies['sum2015'][0]
supplies2016 = parks_supplies['sum2016'][0]
supplies2017 = parks_supplies['sum2017'][0]
supplies2018 = parks_supplies['sum2018'][0]
```

```python
parks_salaries = pd.read_sql("""select * , sum(year2015) as sum2015, sum(year2016) as sum2016,
sum(year2017) as sum2017, sum(year2018) as sum2018 from cpe_table
where Character_Code LIKE '%salaries'

""", cpe_conn)

salaries2015 = parks_salaries['sum2015'][0]
salaries2016 = parks_salaries['sum2016'][0]
salaries2017 = parks_salaries['sum2017'][0]
salaries2018 = parks_salaries['sum2018'][0]
```

```python
parks_benefits = pd.read_sql("""select * , sum(year2015) as sum2015, sum(year2016) as sum2016,
sum(year2017) as sum2017, sum(year2018) as sum2018 from cpe_table
where Character_Code LIKE '%benefits'

""", cpe_conn)
benefits2015 = parks_benefits['sum2015'][0]
benefits2016 = parks_benefits['sum2016'][0]
benefits2017 = parks_benefits['sum2017'][0]
benefits2018 = parks_benefits['sum2018'][0]
```

```python
parks_interDeptCharges = pd.read_sql("""select * , sum(year2015) as sum2015, sum(year2016) as sum2016,
sum(year2017) as sum2017, sum(year2018) as sum2018 from cpe_table
where Character_Code LIKE '%inter depart charges'

""", cpe_conn)
interDeptChrg2015 = parks_interDeptCharges['sum2015'][0]
interDeptChrg2016 = parks_interDeptCharges['sum2016'][0]
interDeptChrg2017 = parks_interDeptCharges['sum2017'][0]
interDeptChrg2018 = parks_interDeptCharges['sum2018'][0]
```

```python
parks_purchasedServices = pd.read_sql("""select * , sum(year2015) as sum2015, sum(year2016) as sum2016,
sum(year2017) as sum2017, sum(year2018) as sum2018 from cpe_table
where Character_Code LIKE '%purchased services'

""", cpe_conn)
purchasedServices2015 = parks_purchasedServices['sum2015'][0]
purchasedServices2016 = parks_purchasedServices['sum2016'][0]
purchasedServices2017 = parks_purchasedServices['sum2017'][0]
purchasedServices2018 = parks_purchasedServices['sum2018'][0]
```
